## üìà **Goals**



-**Role of EDA:** To systematically examine the data in order to understand its behavior, uncover hidden patterns, identify meaningful relationships between variables, and detect unusual or unexpected values that could impact analysis and conclusions.



-**Role of Visualization:** To visually represent data in a clear and intuitive manner, enabling quicker recognition of trends, comparisons, and correlations, and allowing insights to be communicated more effectively than through numerical summaries alone.

## üìà **Analysis Approach:**



 1. **Univariate Analysis**   ‚Üí  Analyzing single variables (distributions, counts, summary statistics).



 2. **Bivariate analysis**    ‚Üí  Studying relationships between two variables (comparisons and correlations)



 3. **Multivariate analysis** ‚Üí  Examining interactions among multiple variables together (e.g., correlation heatmaps)

## üìà **Analysis Focus Areas**



- Understanding overall student academic performance through the distribution of final grades.



- Examining study-related factors such as study time, past failures, and absences to understand their impact on performance.



- Analyzing the relationship between earlier academic scores (G1 and G2) and final outcomes (G3).



- Comparing student performance across different demographic groups such as gender and family background.



- Evaluating the influence of parental education and support-related factors on student achievement.



- Identifying patterns that indicate students at risk of poor academic performance.



- Exploring correlations among academic, behavioral, and social factors to uncover key drivers of success.



- Assessing the impact of school-related support systems such as extra classes and educational assistance on student performance.



- Examining how lifestyle and social factors, including family relationships and free-time activities, influence academic outcomes.



- Analyzing the effect of health status and daily routines on students‚Äô ability to perform academically.

## üìà **Visualization Tools to be Used:**



- Matplotlib : Static visualizations



- Seaborn : Statistical visualizations

```python
# Import matplotlib and seaborn for data visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style for all plots
sns.set(style="whitegrid")


# Display setting for Visualization
sns.set_context("notebook")
df = pd.read_csv(r"C:\Users\anton\Downloads\archive\Student_Performance_Cleaned.csv")
```

## üìà **Distribution of Final Grades (Univariate Analysis)**

This visualization shows how final grades are distributed among students.

It helps in understanding overall academic performance and identifying common score ranges.

```python
plt.figure()
sns.histplot(df['G3'], bins=20, kde=True)
plt.title("Distribution of Final Grade (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

### üîç **# Insight & Recommendation:**



 - Most students‚Äô final grades (G3) are concentrated between 10 and 15, indicating average to moderately good academic performance across the dataset.



 - Very few students score extremely low (0‚Äì5) or very high (18‚Äì20), showing limited cases at both performance extremes.



 - The distribution is slightly right-skewed, suggesting that while most students pass, top academic excellence is less common.



Recommendation: Academic interventions should focus on lifting mid-performing students into higher performance bands, while targeted support programs should be designed for low-performing students to reduce failure risk.

## üìà **Gender Distribution**

This analysis shows the number of students by gender in the dataset.

It helps understand whether the dataset is balanced across genders.

```python
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

### üîç **# Insight & Recommendation:**



- The dataset contains more female students than male students, indicating a gender imbalance in the sample.



- Female students form the majority of the population, which may influence overall performance trends observed in the analysis.



- The difference suggests that conclusions drawn from the dataset may be more representative of female students‚Äô academic behavior.



Recommendation: Gender-based performance comparisons should be interpreted carefully, and further analysis should examine whether academic outcomes differ by gender to ensure balanced and fair insights

## üìà **Study Time Distribution**

This chart visualizes how much time students spend studying.

Study time is an important factor that may influence academic performance.

```python
plt.figure()
sns.countplot(x='studytime', data=df)
plt.title("Study Time Distribution")
plt.xlabel("Study Time (1 = Low, 4 = High)")
plt.ylabel("Number of Students")
plt.show()
```

### üîç **# Insight & Recommendation:**



- The majority of students fall under study time levels 1 and 2, indicating that most students spend low to moderate time studying.



- Relatively fewer students are in higher study time categories (levels 3 and 4), suggesting limited engagement in extended study hours.



- This imbalance indicates that increased study time is not common among most students, which may impact overall academic performance.



Recommendation: Institutions and educators should encourage better study habits and time management strategies, especially for students in lower study time categories, to help improve academic outcomes.

## üìà **Study Time vs Final Grade (Bivariate Analysis)**

This comparison examines whether increased study time is associated with higher final grades.

Box plots help compare grade distributions across different study time levels.

```python
plt.figure()
sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

### üîç **# Insight & Recommendation:**



- Students with higher study time levels (3 and 4) generally show higher median final grades (G3) compared to those with lower study time levels (1 and 2).



- Lower study time categories exhibit greater variability and more low-score outliers, indicating inconsistent academic performance.



- The trend suggests a positive relationship between study time and academic achievement, although some overlap exists across categories.



Recommendation: Encourage students to increase effective study time through structured study plans and academic guidance, particularly targeting students in lower study time groups to improve overall performance.

## üìà **Absences vs Final Grade**

This scatter plot analyzes the relationship between student absences and final grades.

It helps identify whether higher absences negatively impact academic performance.

```python
plt.figure()
sns.scatterplot(x='absences', y='G3', data=df)
plt.title("Absences vs Final Grade")
plt.xlabel("Number of Absences")
plt.ylabel("Final Grade")
plt.show()
```

### üîç **# Insight & Recommendation:**



- Students with fewer absences tend to achieve higher and more consistent final grades, while students with higher absences show greater variability in performance.



- As the number of absences increases, final grades generally tend to decline, indicating a negative relationship between attendance and academic performance.



- A few students with high absences still achieve reasonable grades, suggesting that other factors may partially offset poor attendance.



Recommendation: Schools should closely monitor attendance patterns and implement early intervention strategies for frequently absent students, as improving attendance is likely to have a positive impact on academic outcomes.

## üìà **Relationship Between G1, G2, and G3**

These plots analyze how grades from earlier periods (G1 and G2) influence the final grade (G3).

Strong relationships indicate predictive academic patterns.

```python
plt.figure()
sns.scatterplot(x='G1', y='G3', data=df)
plt.title("G1 vs G3")
plt.xlabel("First Period Grade (G1)")
plt.ylabel("Final Grade (G3)")
plt.show()

plt.figure()
sns.scatterplot(x='G2', y='G3', data=df)
plt.title("G2 vs G3")
plt.xlabel("Second Period Grade (G2)")
plt.ylabel("Final Grade (G3)")
plt.show()
```

### üîç **# Insight & Recommendation (G1 vs G3 & G2 vs G3):**



- Both first period grades (G1) and second period grades (G2) show a strong positive relationship with the final grade (G3), indicating consistent academic performance across assessment periods.



- The relationship between G2 and G3 appears slightly stronger than G1 and G3, suggesting that more recent academic performance is a better predictor of final outcomes.



- Students with low G1 or G2 scores are more likely to achieve low final grades, highlighting early warning signs of academic risk.



- A small number of students with moderate early grades still end up with very low final scores, indicating possible disengagement or external factors affecting performance.



Recommendation: Educational institutions should use G1 and especially G2 scores as early indicators to identify at-risk students and implement timely academic support, mentoring, or remedial programs to improve final performance

## üìà **Correlation Analysis (Multivariate Analysis)**

A correlation heatmap is used to understand relationships among multiple numerical variables.

This helps identify which factors are most strongly related to final academic performance.

```python
numeric_cols = ['G1', 'G2', 'G3', 'studytime', 'absences', 'failures']
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Variables")
plt.show()
```

### üîç **# Insight & Recommendation:**



- There is a very strong positive correlation between academic period grades, especially between G2 and G3, indicating that recent academic performance is the strongest predictor of final grades.



- G1 and G2 also show a strong correlation, suggesting consistent performance patterns across assessment periods.



- Study time has a weak to moderate positive correlation with final grades, indicating that while studying more helps, it is not the only factor influencing performance.



- Absences show a negative correlation with grades, meaning increased absenteeism generally leads to lower academic performance.



- Past failures have a moderate negative correlation with all grade variables, highlighting repeated failures as a key risk factor for poor final outcomes.



Recommendation: Academic monitoring systems should prioritize G2 scores and failure history to identify at-risk students early, while interventions should focus on improving attendance and effective study habits to support better academic results.
